```python
import numpy
import matplotlib.pyplot as plt


def MF_SGD(R1, R2, P, Q, steps, alpha, beta):
    Q = Q.T
    counter = 0
    while True:
        for i in range(len(R1)):
            for j in range(len(R1[0])):
                if R1[i][j] > 0:
                    eij = R1[i][j] - numpy.dot(P[i,:],Q[:,j])
                    P[i,:] = P[i,:] + alpha * (eij * Q[:,j] - beta * P[i,:])
                    Q[:,j] = Q[:,j] + alpha * (eij * P[i,:] - beta * Q[:,j])
        error0 = 0
        error1 = 0
        for i in range(len(R1)):
            for j in range(len(R1[i])):
                if R1[i][j] > 0:
                    error0 += pow(R1[i][j] - numpy.dot(P[i,:],Q[:,j]), 2)\
                              +(beta/2) * numpy.dot(P[i,:],P[i,:])\
                              +(beta/2) * numpy.dot(Q[:,j],Q[:,j])
        for i in range(len(R2)):
            for j in range(len(R2[i])):
                if R2[i][j] > 0:
                    error1 += pow(R2[i][j] - numpy.dot(P[i,:],Q[:,j]), 2)\
                              + (beta/2) * numpy.dot(P[i,:], P[i,:])\
                              + (beta/2) * numpy.dot(Q[:,j], Q[:,j])
        print(counter,' ', error0,' ', error1)
        reg1.append(counter)
        reg2.append(error0)
        reg3.append(error1)
        counter += 1
        if counter > steps:
            plt.plot(reg1, reg2, label='training loss')
            plt.plot(reg1, reg3, label='validation loss')
            plt.xlabel('Iteration')
            plt.ylabel('Loss')
            plt.legend()
            plt.show()
            break
    return P, Q.T

def GetR(filename):
    R = [[0 for col in range(1682)] for row in range(943)]
    for line in open(filename, 'r'):
        (userid, movieid, rating, timestamp) = line.split('\t')
        R[int(userid)-1][int(movieid)-1] = float(rating)
    return R


if __name__ == "__main__":

    R_train = GetR('./ml-100k/u1.base')
    R_val = GetR('./ml-100k/u1.test')

    R_train = numpy.array(R_train)
    R_val = numpy.array(R_val)

    N = len(R_train)
    M = len(R_train[0])
    K = 2
    steps = 1000
    alpha = 0.001
    beta = 0.02

    reg1 = []
    reg2 = []
    reg3 = []

    P = numpy.random.rand(N, K)
    Q = numpy.random.rand(M, K)

    nP, nQ = MF_SGD(R_train, R_val, P, Q, steps, alpha, beta)
    nR = numpy.dot(nP, nQ.T)
    numpy.save("result.npy", nR)
```
